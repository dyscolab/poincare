```python
from symbolite.abstract.lang import Block, Assign
from symbolite.ops import translate
from symbolite.abstract import Real
from symbolite.impl import libpythoncode
```

```python
x = Real("x")
y = Real("y")
b = Block(inputs=(x,), outputs=(y,), lines=(Assign(y, 2 * x),), name="step")
```

```python
translate(b, libpythoncode)
```
